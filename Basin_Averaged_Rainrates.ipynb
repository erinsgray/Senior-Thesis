```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
```

```python
top2016 = pd.read_csv("2016_BasinAverageExcel/Top_BasinAverageRR.csv", header = None, names = ['dates', 'mm'], parse_dates = ['dates'], index_col=['dates'])
bottom2016 = pd.read_csv("2016_BasinAverageExcel/Bottom_BasinAverageRR.csv", header = None, names = ['dates', 'mm'], parse_dates = ['dates'], index_col=['dates'])
tiberrun2016 = pd.read_csv("2016_BasinAverageExcel/TiberRun_BasinAverageRR.csv", header = None, names = ['dates', 'mm'], parse_dates = ['dates'], index_col=['dates'])
top2018 = pd.read_csv("2018_BasinAverageExcel/Top_BasinAverageRR.csv", header = None, names = ['dates', 'mm'], parse_dates = ['dates'], index_col=['dates'])
bottom2018 = pd.read_csv("2018_BasinAverageExcel/Bottom_BasinAverageRR.csv", header = None, names = ['dates', 'mm'], parse_dates = ['dates'], index_col=['dates'])
tiberrun2018 = pd.read_csv("2018_BasinAverageExcel/TiberRun_BasinAverageRR.csv", header = None, names = ['dates', 'mm'], parse_dates = ['dates'], index_col=['dates'])
tiber_area = 1.360033623813821 # actual area
bottom_area = 3.962489282340931 # actual area
top_area = 3.9904212946017354 # actual area
```

```python
## finding and showing actual values
divnum = (60/10.8*4) # timing, 60 minutes per hour with measurements every 10.8 minutes
mmtop2016 = np.trapz(top2016['mm'])/divnum
mmbottom2016 =  np.trapz(bottom2016['mm'])/divnum
mmtiberrun2016 = np.trapz(tiberrun2016['mm'])/divnum
print('2016 Sucker-Hudson = ', mmtop2016, ' mm')
print('2016 Tiber-Bonnie = ', mmbottom2016, ' mm')
print('2016 Hudson-Tiber  = ', mmtiberrun2016, ' mm')
print('2016 Sucker-Hudson  = ', mmtop2016/(top_area), 'mm/km2')
print('2016 Tiber-Bonnie = ', mmbottom2016/(bottom_area), 'mm/km2')
print('2016 Hudson-Tiber = ', mmtiberrun2016/(tiber_area), 'mm/km2')
```

```python
to_inch = 0.0393701
intop2016 = mmtop2016 * to_inch
inbottom2016 = mmbottom2016 * to_inch
intiberrun2016 = mmtiberrun2016 * to_inch
print('2016 top = ', intop2016, ' in')
print('2016 bottom = ', inbottom2016, ' in')
print('2016 Tiber Run = ', intiberrun2016, ' in')
p2016 = intop2016 + inbottom2016 +intiberrun2016
print('2016 top percent = ', intop2016/p2016* 100, '%')
print('2016 bottom percent = ', inbottom2016/p2016* 100, '%')
print('2016 Tiber Run percent = ', intiberrun2016/p2016 * 100, '%')
print('2016 top  = ', intop2016/top_area, 'mm/km')
print('2016 bottom = ', inbottom2016/bottom_area, 'mm/km')
print('2016 Tiber Run = ', intiberrun2016/tiber_area, 'mm/km')
```
